```python
import pandas as pd

df = pd.DataFrame({
    "Order Date":["2026-01-05","2026-01-10","2026-02-15","2026-02-20","2026-03-05","2026-03-12","2026-04-08","2026-04-25","2026-05-15","2026-05-28"],
    "Product":["Laptop","Mobile","Headphones","Smartwatch","Laptop","Mobile","Headphones","Smartwatch","Laptop","Mobile"],
    "Quantity":[2,5,10,4,1,3,8,6,3,7],
    "Price":[50000,15000,2000,5000,50000,15000,2000,5000,50000,15000]
})

df
```

```python
print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]

df
```

```python
import matplotlib.pyplot as plt

revenue_by_product = df.groupby("Product")["Revenue"].sum()

revenue_by_product.plot(kind="barh")

plt.title("Product-wise Revenue")
plt.xlabel("Revenue")
plt.ylabel("Product")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df["Revenue"] = df["Quantity"] * df["Price"]

df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
monthly_revenue = df.groupby(df["Order Date"].dt.month)["Revenue"].sum()

print(monthly_revenue)
```

```python
plt.plot(monthly_revenue.index, monthly_revenue.values, marker="o")

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Quantity"], df["Revenue"])

plt.title("Quantity vs Revenue")
plt.xlabel("Quantity")
plt.ylabel("Revenue")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df["Revenue"], bins=5)

plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")

plt.show()
```
